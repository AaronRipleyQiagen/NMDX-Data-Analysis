```python
##Step 1 Import nmdatalytix into workbook
import nmdatalytix as nmdx
```

```python
##Step 2 "Scape Files" your files (found in a folder) into the Database
##Note scraper will first identify folders within the directory provided,
##with this said please make sure your raw data files are one folder deep in the directory you are pointing to.
nmdx.scarpeFiles("P:/USERS/Aaron/Python Data Analysis Tools/nmdatalytix","test")
```

```python
##Step 3 Retrieve these files from the database by using the tag you just assigned.
exampleDataFrame = nmdx.getneumodxRawDataDBDataAll(environment_list=['test'])
```

```python
##Step 4: Do some cool stuff with your dataframe

##Add consumable lot columns
exampleDataFrame = nmdx.retrieveConsumableLots(exampleDataFrame)
##Add consumable serial number columns
exampleDataFrame = nmdx.retrieveConsumableSerials(exampleDataFrame)
##Add Run Numbers columns
exampleDataFrame = nmdx.getRunNumbers(exampleDataFrame)
display(exampleDataFrame)
```

```python
#Step 5: Summarize your Data
nmdx.getRunStats(exampleDataFrame[exampleDataFrame['Processing Step']=='Raw'], statList=['Ct'])
```

```python
##Step 6: Plot Your Data
plot = nmdx.amplificationPlot(exampleDataFrame, colField='Target Name',
                       rowField='Processing Step',
                       row_order=['Normalized', 'Raw', '2nd'],
                       col_order=['FluA', 'FluB', 'SARS-CoV-2', 'RSV', 'SPC2'], plotwidth=30)
```
