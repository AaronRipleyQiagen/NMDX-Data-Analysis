```python
# root = tk.Tk()
# root.withdraw()

# file = filedialog.askopenfilename()
#df = pd.read_csv(file)
df = pd.read_csv('Gen2_flat_data_sample.csv')
```

```python
dfm = df.copy()
dfm = dfm[dfm.ProcessingStep == "Normalized"]
pd.set_option("display.max_columns", None)
dfm.head(10)
```

```python
l=dfm['Channel'].tolist()
np.unique(l)
```

```python
dfg = dfm[dfm.Channel == "Green_470_510"]
dfy = dfm[dfm.Channel == "Yellow_530_555"]
dfo = dfm[dfm.Channel == "Orange_585_610"]
dffr = dfm[dfm.Channel == "Far_Red_680_715"]
dfr = dfm[dfm.Channel == "Red_625_660"]
```

```python
dfg.head(5)
```

```python
dfg= dfg.filter(items= ["SampleUId","SampleID","ResultCode",
                    "AssayVersion",
                    'N500SerialNumber','NeuMoDxSoftwareVersion','PcrModuleSerial',
                    'CartridgeLot','CartridgeExpirationDate',
                    'BufferTroughLot','BufferTroughExpirationDate',
                    'ExtractionPlateLot','ExtractionPlateExpirationDate',
                    'TestStripLot','TestStripExpirationDate',
                    'ReleaseReagentLot','ReleaseReagentExpirationDate',
                    'WashReagentLot','WashReagentExpirationDate',
                    "TargetName","LocalizedResult",
                    "Ct","EndPointFluorescence","MaxPeakHeight","EPR"])
dfg = dfg.rename(columns= {"TargetName":"Green TargetName","LocalizedResult":"Green LocalizedResult","Ct":"Green Ct","EndPointFluorescence":"Green EP","MaxPeakHeight":"Green MPH","EPR":"Green EPR"})

dfy= dfy.filter(items= ["SampleUId","TargetName","LocalizedResult",
                    "Ct","EndPointFluorescence","MaxPeakHeight","EPR"])
dfy = dfy.rename(columns= {"TargetName":"Yellow TargetName","LocalizedResult":"Yellow LocalizedResult","Ct":"Yellow Ct","EndPointFluorescence":"Yellow EP","MaxPeakHeight":"Yellow MPH","EPR":"Yellow EPR"})  
            
dfo= dfo.filter(items= ["SampleUId","TargetName","LocalizedResult",
                    "Ct","EndPointFluorescence","MaxPeakHeight","EPR"])
dfo = dfo.rename(columns= {"TargetName":"Orange TargetName","LocalizedResult":"Orange LocalizedResult","Ct":"Orange Ct","EndPointFluorescence":"Orange EP","MaxPeakHeight":"Orange MPH","EPR":"Orange EPR"})

dffr= dffr.filter(items= ["SampleUId","TargetName","LocalizedResult",
                    "Ct","EndPointFluorescence","MaxPeakHeight","EPR"])
dffr = dffr.rename(columns= {"TargetName":"FarRed TargetName","LocalizedResult":"FarRed LocalizedResult","Ct":"FarRed Ct","EndPointFluorescence":"FarRed EP","MaxPeakHeight":"FarRed MPH","EPR":"FarRed EPR"})

dfr= dfr.filter(items= ["SampleUId","TargetName","LocalizedResult",
                    "Ct","EndPointFluorescence","MaxPeakHeight","EPR"])
dfr = dfr.rename(columns= {"TargetName":"Red TargetName","LocalizedResult":"Red LocalizedResult","Ct":"Red Ct","EndPointFluorescence":"Red EP","MaxPeakHeight":"Red MPH","EPR":"Red EPR"})
```

```python
dfg
```

```python
dfmg = pd.merge(dfg, dfy, on = ['SampleUId'])
dfmg = pd.merge(dfmg, dfo, on = ['SampleUId'])
dfmg = pd.merge(dfmg, dffr, on = ['SampleUId'])
dfmg = pd.merge(dfmg, dfr, on = ['SampleUId'])
dfmg.head(2)
```

```python
dfline = dfmg.copy()
dfline = dfline.replace("TargetAmplified", "AMP")
dfline = dfline.replace("TargetNotAmplified", "NoAMP")
dfline = dfline.replace("NoResult", "NR")
dfline = dfline.replace("TargetUnresolved", "UNR")
dfline = dfline.replace("Unresolved", "UNR")
dfline = dfline.replace("TargetIndeterminate", "IND")
dfline = dfline.replace("Indeterminate", "IND")
```

```python
dfline.to_csv('Line_Comtest.csv', encoding='utf-8', index=False)
```
